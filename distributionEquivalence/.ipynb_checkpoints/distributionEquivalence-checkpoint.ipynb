```python
import numpy as np
import pdb
import itertools
```

```python
def applyDistPerm(distPerm, distArray):
    
    varPerm = distPerm['varPerm']
    valuePermList = distPerm['valuePermList']
    
    numVar = len(varPerm)
    numValue = len(valuePermList[0])
    
    outArray = np.copy(np.transpose(distArray, varPerm))
    
    for j in range(numVar):
        thisValuePerm = valuePermList[j]
        tempArray = np.copy(outArray)
        for i in range(numValue):
            thisDest = thisValuePerm[i]
            
            destTuple = [slice(0,numValue) for k in range(numVar)]
            destTuple[j] = thisDest
            destTuple = tuple(destTuple)
            
            sourceTuple = [slice(0,numValue) for k in range(numVar)]
            sourceTuple[j] = i
            sourceTuple = tuple(sourceTuple)
            
            outArray[destTuple] = tempArray[sourceTuple]
        
    return outArray

def distEquivCheck(distArray1, distArray2):
    numVar = len(distArray1.shape)
    numValue = distArray1.shape[0]
    
    allVarPerm = list(itertools.permutations(range(numVar)))
    allValuePerm = list(itertools.permutations(range(numValue)))
    allProduct = list(itertools.product(allValuePerm, repeat=numVar))
    counter = 0
    
    for thisVarPerm in allVarPerm:
        for thisTuple in allProduct:
            counter += 1
            distPerm = {'varPerm': thisVarPerm, 'valuePermList': thisTuple}
            permutedDistribution = applyDistPerm(distPerm,distArray1)
            if np.array_equal(distArray2, permutedDistribution):
                return distPerm
    
    print "permutations checked: " + str(counter)
    return False

```
