```python
%matplotlib notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
plt.style.use('ggplot')
```

```python
df = pd.read_csv("C:\\Users\\the_m\\github\\python\\eeg\\julia\\kuramoto_Out_01-19-11-52.csv")
odePhi = df.values[:,1:].transpose()
t = df.values[:,0]
```

```python
plt.figure()
for i in range(100):
    plt.plot(t, odePhi[i,:])
plt.title("Phase Evolution")
```

```python
orderParameterAbs = [abs(sum(np.exp(odePhi[:, i] * (0 + 1j)))) for i in range(len(t))]

plt.figure()
plt.plot(t, orderParameterAbs);
plt.title("Order Parameter Evolution")
```

```python
plt.figure()
thetaFinal = odePhi[:,-1] % (2 * np.pi)
plt.plot(np.cos(thetaFinal), np.sin(thetaFinal),'.')
plt.title("Phases at final time")
```
