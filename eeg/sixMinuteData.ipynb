```python
%matplotlib

import pandas as pd
import numpy as np
import mne
```

```python
dataPath = '/Users/the_m/Google Drive/NeuroResearch/data/NewMCPT1_wake1.txt'
#dataPath = '/Users/mettinger/Google Drive/NeuroResearch/data/NewMCPT1_sleep1.txt'
```

```python
with open(dataPath)as fp:
    lines = fp.readlines()
    
columns = lines[1].strip().split()
data = np.array([lines[i].strip().split() for i in range(2,len(lines))])
print(lines[0])

#df = pd.DataFrame(data=data, columns=columns)
#df.head()

sfreq=1000./float(lines[0].split("SamplingInterval[ms]=")[1].split(' ')[0])

info = mne.create_info(ch_names=columns, sfreq=sfreq)
raw = mne.io.RawArray(data=data.T, info=info)
```

```python
raw.plot(duration=1)
```
