```python
%matplotlib 

import mne
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat("../../NeuroPattToolbox/sampleData_marmosetLFP.mat")
fs = data['sampleFs'][0][0]
lfp = data['sampleLFP']
lfp.shape
```

```python
rawArray = lfp[:,:,:,0].reshape((100,2035))
#rawArray = lfp[0,:,:,0]

info = mne.create_info(ch_names=rawArray.shape[0], sfreq=fs)
raw = mne.io.RawArray(data=rawArray, info=info)
```
