```python
# generate correlation matrix dim x dim
def randomCorrelationMatrix(dim):
    eigs = np.random.randint(low=1, high=10, size=dim)
    eigs = dim * eigs/sum(eigs)
    return random_correlation.rvs(eigs)

    
    
# calculate filtration for a correlation matrix
def filtration(correlationMatrix):
    
    cutoffs = np.triu(correlationMatrix,1).flatten()
    cutoffs = cutoffs[cutoffs != 0]
    cutoffs = np.sort(cutoffs)[::-1] # sort in descending order

    rows,cols = correlationMatrix.shape
    g = nx.Graph()
    
    stream = api.Plex4.createExplicitSimplexStream(1)
    graphList = []
    
    for i in range(rows):
        stream.addVertex(i, cutoffs[0])
        g.add_node(i)
        
    for cutoffIndex,thisCutoff in enumerate(cutoffs):
        for i in range(1, rows):
            for j in range(0, i):
                if np.isclose(correlationMatrix[i,j] , thisCutoff):
                    g.add_edge(i,j)
                    
        graphList.append(g.copy())
        
        cliques = list(nx.algorithms.find_cliques(g))
        cliques = [[int(j) for j in i] for i in cliques]
        
        for thisClique in cliques:
            stream.addElement(thisClique, thisCutoff)

        stream.ensureAllFaces();

    stream.finalizeStream();
    
    return stream, graphList

# plot the list of networkx graphs
def plotGraphs(graphList):
    plt.figure()
    for i in range(len(graphList)):
        plt.subplot(1,len(graphList), i+1)
        nx.draw(graphList[i])
    
# calculate the persistence diagram for a correlation matrix
def matrixToPersistencePoints(matrix, topHomology=100, coefficientField=2):
    filt, graphList = filtration(matrix)
    persistence = api.Plex4.getModularSimplicialAlgorithm(topHomology, coefficientField);
    intervals = persistence.computeIntervals(filt)
    dimensionList = [thisDimension.getValue() for thisDimension in intervals.getIntervalIterator()]
    intervals = [[(thisInterval.getStart(), thisInterval.getEnd()) for thisInterval in thisList] for thisList in dimensionList]
    return intervals

# plot a persistence diagram
def plotPersistence(intervals, axis=None):
    if axis==None:
        fig = plt.figure()
        axis = fig.add_subplot(111)
        
    for i,thisDimension in enumerate(intervals):
        x = [i[0] for i in thisDimension]
        y = [i[1] if i[1] != None else i[0] for i in thisDimension]
        axis.scatter(x,y,marker='o')
        
# create 3d cube of interpolated matrices
def matrixInterp(m1, m2, t_interp):
    cube = np.zeros(shape= list(m1.shape) + [len(t_interp)])
    for i, this_t in enumerate(t_interp):
        cube[:,:,i] = m1 + (m2-m1) * this_t
    return cube

def temporalPersistence(correlationCube):
    
    def animate(i):
        ax.clear()
        plotPersistence(allIntervals[i], ax)
        return ax

    depthFrames=200
    intervalMillisec = 100

    rows, cols, depth = correlationCube.shape
    allIntervals = [matrixToPersistencePoints(correlationCube[:,:,i]) for i in range(depth)]
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    anim = FuncAnimation(fig, animate, frames=depth, interval=intervalMillisec, repeat=False)
    return anim
```

```python
corr_1 = randomCorrelationMatrix(10)
corr_2 = randomCorrelationMatrix(10)
```

```python
correlationCube = matrixInterp(corr_1, corr_2, np.linspace(0,1,100))
anim = temporalPersistence(correlationCube)
```
