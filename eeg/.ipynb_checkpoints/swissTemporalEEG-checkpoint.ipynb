```python
%matplotlib 

import h5py
import mne
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
```

# Load hdf5 file

```python
subject = '04'
session = '01'

filename = '/Users/the_m/Desktop/swissEEG/data_nix/Data_Subject_%s_Session_%s.h5' % (subject, session)
f = h5py.File(filename, 'r')

print(f.keys())
print(f['metadata'].keys())
print(f['data'].keys())
```

# Get datasets in hdf5 file

```python
def visitor_func(name, node):
    if isinstance(node, h5py.Dataset):
         datasetList.append(node.name)
    else:
         groupList.append(node.name)
            
datasetList = []
groupList = []
f.visititems(visitor_func)

data = [i for i in datasetList if i.startswith('/data')]
metadata = [i for i in datasetList if i.startswith('/metadata')]
```

# Data names

```python
for i in data:
    print(i)
```

# Metadata names

```python
for i in metadata:
    print(i)
```

# Plot electrode coordinates in 3d

```python
key = '/data/Data_Subject_%s_Session_%s/data_arrays/Scalp_Electrode_EEGLAB_BESA_Coordinates/data' % (subject, session)
scalpCoordinates = f[key][:]

key = '/data/Data_Subject_%s_Session_%s/data_arrays/iEEG_Electrode_MNI_Coordinates/data' % (subject, session)
depthCoordinates = f[key][:]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xDepth = depthCoordinates[:,0]
yDepth = depthCoordinates[:,1]
zDepth = depthCoordinates[:,2]

ax.scatter(xDepth, yDepth, zDepth, c='b', marker='o')

xScalp = scalpCoordinates[:,0]
yScalp = scalpCoordinates[:,1]
zScalp = scalpCoordinates[:,2]

ax.scatter(xScalp, yScalp, zScalp, c='r', marker='o')

plt.show()
```

# Load trial data for iEEG or scalp EEG

```python
trialNum = '06'
scalp = False

if scalp:
    trialString = '/data/Data_Subject_%s_Session_%s/data_arrays/Scalp_EEG_Data_Trial_%s' % (subject, session, trialNum)
    sfreq = 200
else:
    trialString = '/data/Data_Subject_%s_Session_%s/data_arrays/iEEG_Data_Trial_%s' % (subject, session, trialNum)
    sfreq = 2000


data = trialString + '/data'
labels = trialString + '/dimensions/1/labels'

rawArray = np.array(f[data])
columns = [i.decode('UTF-8') for i in list(f[labels])]

info = mne.create_info(ch_names=columns, sfreq=sfreq)
raw = mne.io.RawArray(data=rawArray, info=info)
```

```python
raw.plot()
```

```python
frames=200
intervalMillisec = 100
contours = 0

vmax = np.max(rawArray)
vmin = -vmax

fig = plt.figure()
def animate(i):
    plt.clf()
    im, cn = mne.viz.plot_topomap(rawArray[:,i],scalpCoordinates[:,0:2], contours=contours, vmin=vmin, vmax=vmax, cmap='RdBu_r')
    plt.title(str(i))
    plt.colorbar(im)
    return im

anim = FuncAnimation(fig, animate, frames=frames, interval=intervalMillisec)

#anim.save('scalp.gif')
```

```python
depthFrames=200
intervalMillisec = 100
depthColormap = 'YlOrRd'
scalpColormap = 'BuGn'
trialNum = '06'

trialStringScalp = '/data/Data_Subject_%s_Session_%s/data_arrays/Scalp_EEG_Data_Trial_%s' % (subject, session, trialNum)
trialStringDepth = '/data/Data_Subject_%s_Session_%s/data_arrays/iEEG_Data_Trial_%s' % (subject, session, trialNum)

rawArrayScalp = np.array(f[trialStringScalp + '/data'])
rawArrayScalp = rawArrayScalp - np.min(rawArrayScalp)
rawArrayScalp = rawArrayScalp/np.max(rawArrayScalp)
scalpMax = np.max(rawArrayScalp)
scalpMin = np.min(rawArrayScalp)

rawArrayDepth = np.array(f[trialStringDepth + '/data'])
rawArrayDepth = rawArrayDepth - np.min(rawArrayDepth)
rawArrayDepth = rawArrayDepth/np.max(rawArrayDepth)
depthMax = np.max(rawArrayDepth)
depthMin = np.min(rawArrayDepth)


vmax = np.max(rawArray)
vmin = -vmax

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
def animate(i):
    ax.scatter(xDepth, yDepth, zDepth, c=rawArrayDepth[:,i], cmap=depthColormap, marker='o',vmin=depthMin, vmax=depthMax)
    ax.scatter(xScalp, yScalp, zScalp, c=rawArrayScalp[:,int(i/10)], cmap=scalpColormap, marker='o', vmin=scalpMin, vmax=scalpMax)
    plt.title(str(i))
    #plt.colorbar(im)
    return ax

anim = FuncAnimation(fig, animate, frames=depthFrames, interval=intervalMillisec, repeat=False)
```

```python
anim.save('animation.gif')
```
