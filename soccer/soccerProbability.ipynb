```python
def oddsCalculate(oddsList):
    probList = np.array([1./i for i in oddsList])
    probList = probList/sum(probList)
    return probList

def outcomeCalculate(rowArray):
    if rowArray[3] == 'H':
        outcomeIndex = 0
    elif rowArray[3] == 'D':
        outcomeIndex = 1
    elif rowArray[3] == 'A':
        outcomeIndex = 2
    else:
        print("Error")
        return
    
    probs = oddsCalculate(rowArray[7:10])
    return outcomeIndex, probs
```
