```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/the_m/Desktop/premierAmerica2021.csv')
```

```python
df.values
```

```python
values2021 = [i for i in df.values if '2021' in i[1]]
```

```python
def processEntry(entry):
    if entry[3].startswith('Withdrawal'):
        entry[3] = entry[3][11:]
        return entry
    else:
        return entry
```

```python
values2021Good = [processEntry(i) for i in values2021]
```

```python
df1 = pd.DataFrame(values2021Good)
df1.columns = df.columns
df1
```

```python
df1.to_csv('C:/Users/the_m/Desktop/premierAmerica2021Good.csv')
```
